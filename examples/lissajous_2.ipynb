```python
import numpy as np
import matplotlib.pyplot as plt

from curve.curves import lissajous
from curve.plot import curveplot
```

```python
def plot_curves(crv1, crv2):
    intersections = crv1.intersect(crv2)
    print(f'The number of intersections: {len(intersections)}')

    intsct_px = [] 
    intsct_py = []

    for n, intersection in enumerate(intersections, start=1):
        p = intersection.intersect_point
        intsct_px.append(p[0])    
        intsct_py.append(p[1])

    f, ax = plt.subplots(1, 1, figsize=(9, 7))

    curveplot(crv1, axes=ax).\
        curveplot(crv2)
    
    st1 = crv1[0]
    st2 = crv2[0]
    
    ax.scatter([st1[0]], [st1[1]], color='r', marker='.', zorder=3)
    ax.scatter([st2[0]], [st2[1]], color='b', marker='.', zorder=4)

    if intsct_px:
        ax.scatter(intsct_px, intsct_py, zorder=5, edgecolor='k', facecolor='r')

def make_curves():
    crv1 = lissajous(t_start=-0.05, t_stop=np.pi*2-0.05, p_count=500, a=3, b=4)
    crv2 = lissajous(t_start=0.05, t_stop=np.pi*2+0.05, p_count=500, a=3, b=4, d=1.0)
    plot_curves(crv1, crv2)
```

```python
make_curves()
```
