## Curves 3D intersections

`scikit-curve` can be used to determine intersections between curves in Euclidean any dimensional space.
This example demonstrates how we can find intersections between two 3D curves.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from curve.curves import helix
from curve.plot import curveplot
```

### Create curve objects

We create two 3D "helix" curves to demonstrate intersections between them.

```python
crv1 = helix(p_count=300)
crv2 = helix(p_count=300, a=-1.0, b=-1.0)
```

### Find intersections

We use `almost` intersection method because in the space >= 3D exact intersections most likely do not exist.
`almost` method uses the algorithm based on shortest segment between two segments of the curves.

```python
%%time
intersections = crv1.intersect(crv2, method='almost', dist_tol=1e-3)
print(f'The number of intersections: {len(intersections)}')
```

### Plot the curves and intersection points

```python
f = plt.figure(figsize=(10, 8))
ax = f.add_subplot(111, projection='3d')

curveplot(crv1, axes=ax).curveplot(crv2)

intsct_px = [] 
intsct_py = []
intsct_pz = []

for n, intersection in enumerate(intersections, start=1):
    p = intersection.intersect_point
    intsct_px.append(p[0])    
    intsct_py.append(p[1])
    intsct_pz.append(p[2])

ax.scatter(intsct_px, intsct_py, intsct_pz, edgecolor='k', facecolor='r', alpha=1.0, s=25);
```
